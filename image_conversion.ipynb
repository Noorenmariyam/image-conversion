## NUMPY GRAYSCALE CONVERSION PROJECT

Objective:

This image processing project aims to convert a color image to grayscale using NumPy. The primary objectives are:

Using NumPy for Image Manipulation: Gain experience with NumPy for manipulating image data effectively.

Applying Mathematical Concepts: Apply concepts like weighted sums to transform RGB values to grayscale.

Preparing for Advanced Topics: Build a foundation for more advanced image processing tasks and computer vision projects.

Overall, the project provides a hands-on approach to learning image processing fundamentals and prepares you for more complex analyses.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the image using PIL
image = Image.open(r"C:\Users\user\Downloads\images.jpeg") 
image_array = np.array(image)

# Display the original image
plt.imshow(image_array)
plt.title("Original Image")
plt.axis('off')
plt.show()

# Step 2: Convert to Grayscale
def rgb_to_grayscale(rgb_image):
    return np.dot(rgb_image[...,:3], [0.2989, 0.5870, 0.1140])

# Apply the grayscale conversion
grayscale_image = rgb_to_grayscale(image_array)

# Display the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```
